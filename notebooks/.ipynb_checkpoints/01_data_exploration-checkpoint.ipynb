```python
import pandas as pd

df = pd.read_csv(r'C:\Users\Lenovo\product-recommendation-system\data\ratings_Beauty.csv')

# Rename columns
df.columns = ['user_id', 'product_id', 'rating', 'timestamp']

# Drop duplicate header row
df = df[df['user_id'] != 'UserId'].reset_index(drop=True)

# Force rating to numeric, any non-numeric becomes NaN
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')

# Drop rows where rating is NaN
df = df.dropna(subset=['rating']).reset_index(drop=True)

print("Shape:", df.shape)
df.head()
```

```python
print("Total users:", df['user_id'].nunique())
print("Total products:", df['product_id'].nunique())
print("Rating range:", df['rating'].min(), "to", df['rating'].max())
print("Any nulls?", df.isnull().sum().sum())
```
